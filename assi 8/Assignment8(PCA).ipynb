```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
# Converting data to numpy array
wine_ary=Wine.values
wine_ary
```

```python
# Normalizing the  numerical data
wine_norm=scale(wine_ary)
wine_norm
```

PCA 

```python
# Applying PCA Fit Transform to dataset
pca = PCA()
pca_values = pca.fit_transform(wine_norm)
pca_values
```

```python
pca.components_
```

```python
# The amount of variance that each PCA has
var = pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
Var = np.cumsum(np.round(var,decimals= 4)*100)
Var
```

```python
plt.plot(Var,color="black");
```

```python
# Final Dataframe
final_df=pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# plot scatterplot
import seaborn as sns
fig=plt.figure(figsize=(20,8))
sns.scatterplot(data1=final_df)
```

Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
#dendrgram
from seaborn.matrix import dendrogram
plt.figure(figsize=(15,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,method='complete'))
```

```python
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hc
```

```python
y=pd.DataFrame(hc.fit_predict(wine_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
data2=wine.copy()
data2['clusterid']=hc.labels_
data2
```

KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
  kmeans=KMeans(n_clusters=i,random_state=2)
  kmeans.fit(wine_norm)
  wcss.append(kmeans.inertia_)
```

```python
#Elbow Graph
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show()
```

Build Cluster Algorithm using k-3

```python
cluster3=KMeans(3,random_state=30).fit(wine_norm)
cluster3
```

```python
cluster3.labels_
```

```python
data3=wine.copy()
data3['cluster3id']=cluster3.labels_
data3
```
