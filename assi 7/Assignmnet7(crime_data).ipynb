```python
#import libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
#load the dataset
crime=pd.read_csv('/content/sample_data/crime_data.csv')
```

```python
crime
```

```python
#Normalization function
#use sklearn library norm
def norm_func(i):
  X=(i-i.min())/(i.max()-i.min())
  return(X)
```

```python
#normalized data frame(considering the numerical part of data)
df_norm=norm_func(crime.iloc[:,1:])
```

```python
df_norm
```

```python
#Create Dendrogram
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='complete'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='average'))
```

```python
#Create Clusters
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='complete')
```

```python
hc.fit(df_norm)
```

```python
#save clusters for chart
y_hc=hc.fit_predict(df_norm)
```

K Means Clustering

```python
from sklearn.cluster import KMeans 
from scipy.spatial.distance import cdist
```

```python
crime1=pd.read_csv('/content/sample_data/crime_data.csv')
```

```python
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
def_norm=norm_func(crime.iloc[:,1:])
```

```python
#elbow curve
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(def_norm)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.title('Elbow curve')
plt.show()
```

```python
#selecting 4 cluster for scree plot
model=KMeans(n_clusters=4)
model.fit(def_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime1['Clust']=x
```

DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime2=pd.read_csv('/content/sample_data/crime_data.csv')
```

```python
crime2.info()
```

```python
df=crime2.iloc[:,1:5]
```

```python
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
pd.concat([crime2,cl],axis=1)
```
