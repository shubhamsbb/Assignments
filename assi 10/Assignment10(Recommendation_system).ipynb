```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
len(book1['Book.Title'].unique())
```

```python
book2=book1.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
book2
```

```python
book2.index=book1['User.ID'].unique()
book2
```

```python
book2.fillna(0,inplace=True)
book2
```

```python
user_sim=1-pairwise_distances(book2.values,metric='cosine')
user_sim
```

```python
user_sim1=pd.DataFrame(user_sim)
user_sim1
```

```python
user_sim1.index=book1['User.ID'].unique()
user_sim1.columns=book1['User.ID'].unique()
user_sim1
```

```python
np.fill_diagonal(user_sim,0)
user_sim1
```

```python
user_sim1.idxmax(axis=1)
```
